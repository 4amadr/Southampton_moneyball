```python
import pandas as pd

df = pd.read_csv('PL_data.csv')
df
```

```python
# Calculate means
desarmes_media = df['Tackles'].mean()
duelos_no_chao_media = df['Ground Duels'].mean()
bolas_zeradas_media = df['Bolas zeradas po 90m'].mean()

# Apply filter comparing each value to the average
defensor_ideal = df[
    (df['Position'] == 'DEF') &
    (df['Tackles'] > desarmes_media) &
    (df['Ground Duels'] > duelos_no_chao_media) &
    (df['Bolas zeradas po 90m'] > bolas_zeradas_media)
]

defensor_ideal.head(40)

defensor_ideal.to_csv('defensor_ideal.csv')
```

```python
# removendo valores de porcentagem
df['fThird Passes %'] = df['fThird Passes %'].str.replace('%', '').astype(float) / 100
```

```python
# identificando alvos para o meio campo
media_assistencias = df['Assists'].mean()
passes_parte_final_media = df['fThird Passes %'].mean()
quebra_de_linha_media_por_jogo = df['Quebra de linha por 90m'].mean()
eficacia_com_bola_media = df['Eficacia com a bola'].mean()
assistencia_por_90m = df['Assistencias por 90m'].mean()
# Clubes de jogadores descartados por conter jogadores superfaturados
clubes_indesejados = ['Arsenal', 'Chelsea', 'Manchester United', 'Manchester City', 'Tottenham Hotspur', 'Liverpool']


meias_ideais = df[
    (~df['Club'].isin(clubes_indesejados)) &
    (df['Position'] == 'MID') &
    (df['Assists'] > media_assistencias) &
    #df['fThird Passes %'] > passes_parte_final_media &
    (df['Quebra de linha por 90m'] > quebra_de_linha_media_por_jogo) &
    (df['Eficacia com a bola'] > eficacia_com_bola_media) &
    (df['Assistencias por 90m'] > assistencia_por_90m)
]
meias_ideais
```

```python
# removendo valores de porcentagem
df['Conversion %'] = df['Conversion %'].str.replace('%', '').astype(float) / 100
```

```python
# analise de atacantes ideais
media_gols = df['Goals'].mean()
chutes_no_gol = df['Shots On Target'].mean()
conversao_porcentagem = df['Conversion %'].mean()
media_gols_por_jogo = df['Gols por 90m'].mean()
chutes_gol_por_jogo = df['Chutes ao gol por 90m'].mean()
gols_vs_assistencias = df['Gols por Assistência'].mean()

atacantes_ideais = df[
    (~df['Club'].isin(clubes_indesejados)) &
    (df['Position'] == 'FWD') &
    (df['Shots On Target'] > chutes_no_gol) &
    (df['Conversion %'] > conversao_porcentagem) &
    (df['Gols por 90m'] > media_gols_por_jogo) &
    (df['Chutes ao gol por 90m'] > chutes_gol_por_jogo) &
    (df['Gols por Assistência'] > gols_vs_assistencias)
]

print(atacantes_ideais)
atacantes_ideais.to_csv('atacantes_filtrados.csv', index=False)
```
